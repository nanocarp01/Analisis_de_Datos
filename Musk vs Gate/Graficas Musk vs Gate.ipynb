```python
import matplotlib.pyplot as plt
import matplotlib.patheffects as pa #libreria para efectos visuales en los graficos
```

```python
import pandas as pd 
```

```python
excel = pd.read_csv(r'D:/pandas/finanza.csv', sep=";")
```

```python
excel.head()
```

```python
excel.dtypes
```

```python
excel['Fecha'] = pd.to_datetime(excel['Fecha'], format = '%d/%m/%Y')
```

```python
excel.head()
```

```python
excel.dtypes
```

```python
excel.set_index('Fecha', inplace = True)
```

```python
excel.head()
```

# Grafico 

```python
plt.style.use("ggplot")

fig, ax = plt.subplots(figsize = (15,8))#dimensionamos el grafico

ax.plot(excel['Elon Mosk'], label = 'Elon Musk', lw=2, color = '#EC4D37',
       path_effects = [pa.SimpleLineShadow(), pa.Normal()])

ax.plot(excel['Bill Gate'], label = 'Bill Gate', lw=2, color = '#1D1B1B', linestyle ="-.",
       path_effects = [pa.SimpleLineShadow(), pa.Normal()])

ax.legend(shadow = True, prop ={"family":"serif", "size":15})

ax.set_ylabel("USD", fontname="serif", fontsize=20)
ax.set_xlabel("Time", fontname="serif", fontsize=20)
ax.set_title("Elon Musk vs Bill Gate", fontname="serif", fontsize=25)

plt.show();
```
